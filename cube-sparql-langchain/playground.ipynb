```python
from langchain.chat_models import ChatOpenAI
from langchain.prompts.chat import ChatPromptTemplate
from langchain.chains import LLMChain
from langchain.schema import HumanMessage

chat_model = ChatOpenAI(openai_api_key=OPENAI_API_KEY, model="gpt-3.5-turbo", temperature=temperature)

structure_description = """
Dimensions labels:
{dimensions_triplets}
"""

sample_description = """
Given cube and its sample observation::
{cube_and_sample}
"""

human_template = "Give me query with prefixes for: {question}"

chat_prompt1 = ChatPromptTemplate.from_messages([
    ("system", sample_description),
    ("system", structure_description),
    ("human", human_template),
])

# rdf_cube = "<https://environment.ld.admin.ch/foen/ubd000502/4>"

# sample_result = """
# @prefix schema: <http://schema.org/> .
# @prefix qudt: <http://qudt.org/schema/qudt/> .
# @prefix sh: <http://www.w3.org/ns/shacl#> .
# @prefix rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#> .
# @prefix rdfs: <http://www.w3.org/2000/01/rdf-schema#> .
# @prefix cube: <https://cube.link/> .

# <https://environment.ld.admin.ch/foen/ubd000502/4/observation/1421/alle%20THG/1999> a cube:Observation ;
#    cube:observedBy <https://ld.admin.ch/office/VII.1.7> ;
#    <https://environment.ld.admin.ch/foen/ubd000502/sektorid> <https://environment.ld.admin.ch/vocabulary/ghg_emission_sectors_co2_ordinance/1421> ;
#    <https://environment.ld.admin.ch/foen/ubd000502/gas> <https://ld.admin.ch/cube/dimension/testdimension/test1> ;
#    <https://environment.ld.admin.ch/foen/ubd000502/jahr> "1999"^^<http://www.w3.org/2001/XMLSchema#gYear> ;
#    <https://environment.ld.admin.ch/foen/ubd000502/werte> 0.588 ;
#    <https://environment.ld.admin.ch/foen/ubd000502/hierachie> "1421" ;
#    <https://environment.ld.admin.ch/foen/ubd000502/werteNichtGerundet> 0.5880504 .
# """

chain1 = LLMChain(prompt=chat_prompt1, llm=chat_model, callbacks=[handler])

ai_template = "{ai_response}"
additional_question = "Filtering by cube {cube} is missing in the query. Please add it to the query."

chat_prompt2 = ChatPromptTemplate.from_messages([
    ("system", sample_description),
    ("system", structure_description),
    ("human", human_template),
    ("ai", ai_template),
    ("human", additional_question),
])
chain2 = LLMChain(prompt=chat_prompt2, llm=chat_model, callbacks=[handler])

question = "Give me query with prefixes for: sum of emission of CO2 for industry between year 2009 and 2011 for given cube <https://environment.ld.admin.ch/foen/ubd000502/4>"

async def acall_chain():
    response = await chain1.ainvoke({
        "dimensions_triplets": dimensions_triplets,
        "cube_and_sample": cube_and_sample,
        "question": question,
    })
    ai_response = response['text']
    cube = "<https://environment.ld.admin.ch/foen/ubd000502/4>"

    if cube not in ai_response:
        response2 = await chain2.ainvoke({
            "dimensions_triplets": dimensions_triplets,
            "cube_and_sample": cube_and_sample,
            "question": question,
            "ai_response": ai_response,
            "cube": cube,
        })

        ai_response = response2['text']

    return ai_response
```

```python
result = await acall_chain()
```
